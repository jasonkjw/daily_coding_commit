```python
#270
class Stock:
  def __init__(self, name, code, PER, PBR, ROE):
    self.name = name
    self.code = code
    self.PER =PER
    self.PBR = PBR
    self.ROE = ROE

  def set_name(self, name):
    self.name = name

  def set_code(self, code):
    self.code = code

  def set_PER(self, PER):
    self.PER = PER

  def set_PBR(self, PBR):
    self.PBR = PBR

  def set_ROE(self, ROE):
    self.ROE = ROE

삼성 = Stock("삼성전자", "005930", 15.79, 1.33, 2.83)
현차 = Stock("현대차","005380", 8.70, 0.35, 4.27)
LG = Stock("LG전자", "066570", 317.34, 0.69, 1.37)

종목 = []

종목.append(삼성)
종목.append(현차)
종목.append(LG)

for i in 종목:
  print(i.name, i.code, i.PER)
```
