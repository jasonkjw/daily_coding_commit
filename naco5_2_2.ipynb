```python
cabinet = {"A-3":"푸", "B-100":"피글렛"}
print(cabinet)
cabinet["A-3"] = "티거"
cabinet["C-20"] = "이요르"
print(cabinet)
del cabinet["A-3"]
print(cabinet)
```

```python
cabinet = {"A-3":"푸", "B-100":"피글렛"}
print(cabinet.keys())
del cabinet["A-3"]
print(cabinet)
print(cabinet.values())
print(cabinet.items())
cabinet.clear()
print(cabinet)
```
