```python
#129
num = input("주민등록번호:")
val1=int(num[0])*2 + int(num[1])*3 + int(num[2])*4 + int(num[3])*5 + int(num[4])*6 \
+int(num[5])*7 + int(num[7])*8 + int(num[8])*9 + int(num[9])*2 + int(num[10])*3 \
+ int(num[11])*4 + int(num[12])*5
val2 = str(11- (val1 % 11))

if num[-1] == val2[-1]:
  print("유효한 주민등록번호입니다.")
else:
  print("유효하지 않은 주민등록번호입니다.")
```

```python
# 130
import requests
btc = requests.get("https://api.bithumb.com/public/ticker/").json()['data']
num1 = int(btc['max_price']) - int(btc['min_price']) # 변동폭
if (int(btc['opening_price']) + num1) > int(btc['max_price']):
  print("상승장")
else :
  print("하락장")
```
