```python
from geopy.distance import geodesic

station_busan = (35.58237, 129.33469) # 울산 관측소
kori_2 = (35.320447, 129.290688)    # 고리 2발

# WGS-84 타원체 기준 거리 계산
dist = geodesic(station_busan, kori_2).kilometers
print(f"거리: {dist:.4f} km")
```

```python
import math
from geopy.distance import geodesic

station_ulsan = (35.58237, 129.33469)
station_elev = 81.74  # m

kori_2 = (35.320447, 129.290688)
kori_elev = 1.7     # m

horizontal_dist = geodesic(station_ulsan, kori_2).kilometers

# EL. (m -> km)
elev_diff_km = abs(station_elev - kori_elev) / 1000

# L = sqrt(d^2 + ΔH^2)
slant_range = math.sqrt(horizontal_dist**2 + elev_diff_km**2)

print(f"1. 수평 거리: {horizontal_dist:.4f} km")
print(f"2. 고도차: {abs(station_elev - kori_elev):.1f} m")
print(f"3. 공간 거리: {slant_range:.4f} km")
```
