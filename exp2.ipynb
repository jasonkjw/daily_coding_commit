```python
# BMI calc.
weight = float(input("몸무게를 kg 단위로 입력하시오:"))
height = float(input("키를 m단위로 입력하시오:"))
BMI = weight / height**2
print(f"당신의 BMI={BMI}")
```

```python
# shpere

radius = float(input("반지름을 입력하시오:"))
vol = 4/3 * 3.14 * radius**3
print(f"구의 부피 = {vol:10.2f}")
```

```python
# atm
price = int(input("물건값을 입력하세요:"))
print("===입력금액===")
bill_1000 = int(input("1000원 지폐개수:"))
coin_500 = int(input("500원 동전개수:"))
coin_100 = int(input("100원 동전개수:"))
cash = bill_1000 * 1000 + coin_500 * 500 + coin_100 * 100
change = cash - price
print("===거스름돈 출력===")
print(f"거스름돈은 {change}원 입니다.")
coin_500 = change // 500
change = change%500
coin_100 = change // 100
change = change%100
coin_10 = change  // 10
change = change%10
coin_1 = change
print(f"500원={change_500}, 100원={change_100}, 10원={change_10}, 1원={change_1}")
```
