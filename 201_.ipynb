```python
# 226
def print_5xn(str):
  for i in range(len(str)//5):
    print(len(str)//5)
    print(i)
    print(str[i:i+5])
    i += 4

print_5xn("아이엠어보이유알어걸")
# print_5xn("12345678912345678912345679")
```
