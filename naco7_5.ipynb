```python
height = float(input("키를 입력하세요(m):"))
gender = str(input("성별을 입력하세요(남자/여자):"))

def std_weight(height, gender):
  if gender == "남자":
    std = height * height * 22
  elif gender == "여자":
    std = height * height * 21
  return std

print(f"키 {height*100}cm 남자의 표준 체중은 {std_weight(height, gender):.2f}kg 입니다.")


```
