```python
import os
from PIL import Image
from google.colab import drive

# 1. 구글 드라이브 마운트 (연결)
drive.mount('/content/drive')

# 2. 경로 설정 (사용자 환경에 맞게 수정하세요)
# 예: 'drive/MyDrive/my_photos' (사진이 들어있는 폴더)
input_folder = 'drive/MyDrive/MT'
output_folder = 'drive/MyDrive/콜라주결과'

if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# 3. 이미지 파일 목록 가져오기 (이름순 정렬)
files = sorted([f for f in os.listdir(input_folder) if f.lower().endswith(('.jpg', '.jpeg', '.png'))])

# 4. 3x3 콜라주 생성 로직
rows, cols = 3, 3
batch_size = rows * cols # 9장씩 끊어서 처리

for i in range(0, len(files), batch_size):
    batch_files = files[i:i + batch_size]
    if not batch_files:
        break

    # 첫 번째 이미지를 기준으로 칸 크기 결정
    first_img = Image.open(os.path.join(input_folder, batch_files[0]))
    w, h = first_img.size

    # 3x3 캔버스 생성
    collage = Image.new('RGB', (w * cols, h * rows), (255, 255, 255))

    for idx, file in enumerate(batch_files):
        img = Image.open(os.path.join(input_folder, file))
        img = img.resize((w, h)) # 모든 사진 크기 통일

        x = (idx % cols) * w
        y = (idx // cols) * h
        collage.paste(img, (x, y))

    # 저장
    collage_name = f"collage_{(i // batch_size) + 1:02d}.jpg"
    collage.save(os.path.join(output_folder, collage_name))
    print(f"{collage_name} 저장 완료!")

print(f"\n총 {len(files)}장의 사진으로 {len(files)//9 if len(files)%9==0 else len(files)//9 + 1}장의 콜라주를 만들었습니다.")
```
