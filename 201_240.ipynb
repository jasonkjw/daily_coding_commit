```python
# 226
def print_5xn(str):
  chk_num = int(len(str)/5)
  for i in range(chk_num + 1):
    print(str[i*5:i*5+5])

print_5xn("아이엠어보이유알어걸")
```

```python
# 227
def print_mxn(string,num):
  ckn = len(string)//num
  for i in range(0,ckn+1):
    print(string[i*num:i*num+ckn])

print_mxn("아이엠어보이유알어걸", 3)

```
