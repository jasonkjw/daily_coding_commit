```python
customer = "토르"
index = 5

while index >= 1:
  print(f"{customer}님, 음료가 준비되었습니다.")
  index -= 1
  print(f"{index}번 남았어요.")
  if index == 0:
    print("커피를 폐기 처분합니다.")
```

```python
customer = "토르"
person = None

while person != customer:
  print(f"{customer}님, 커피가 준비되었습니다.")
  person = input("이름이 어떻게 되세요?")
```
