```python
glasses = 10

def rent(people):
  global glasses
  glasses = glasses - people
  print(f"[함수 내부] 남은 3D 안경 개수: {glasses}")

print(f"전체 3D 안경 개수: {glasses}")
rent(2)
print(f"남은 3D 안경 개수: {glasses}")
```

```python
glasses = 10

def rent_return(glasses, people):
  glasses = glasses - people
  print(f"[함수 내부] 남은 3D 안경 개수:")
  return glasses

print(f"전체 3D 안경 개수: {glasses}")
glasses = rent_return(glasses, 2)
print(f"남은 3D 안경 개수: {glasses}")
```
